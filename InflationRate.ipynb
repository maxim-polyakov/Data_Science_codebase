```python
df = pd.read_csv('../DataSets/Inflation/ADO April 2023 - Inflation.csv', encoding='cp1252')
df.head()
```

```python
print(df.RegionalMember.unique())
df[pd.isna(df['RegionalMember'])]
```

```python
print(df.Year.unique())
print(df[pd.isna(df['Inflation'])])
inf_nan = df[(df[' Country Code'] == 'NIU')|(df[' Country Code'] == 'AFG')].index
df = df.drop(inf_nan)
df = df.reset_index(drop=True)
```

```python
print(df['Unit of Measurement'].unique())
df = df.drop(columns=['Unit of Measurement'])
df
```

```python
object_cols = ['Year']
ordinal_encoder = OrdinalEncoder()
df[object_cols] = ordinal_encoder.fit_transform(df[object_cols])
df
```

```python
object_cols = ['RegionalMember']
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols = pd.DataFrame(OH_encoder.fit_transform(df[object_cols]))
OH_cols.index = df.index

df = df.drop(object_cols, axis=1)
df = pd.concat([df, OH_cols], axis=1)
df.columns = df.columns.astype(str)
#print(df.to_markdown())
```

```python
Country_Code =[]
null_index = []
for i in range(len(df.columns)-4):
    string = ''
    if type(df.loc[df[df.columns[i+4]]>0,' Country Code'].iloc[0]) == str:
        Country_Code.append(df.loc[df[df.columns[i+4]]>0,' Country Code'].iloc[0])
    elif type(df.loc[df[df.columns[i+4]]>0,' Subregion'].iloc[0]) == str:
        Country_Code.append(df.loc[df[df.columns[i+4]]>0,' Subregion'].iloc[0])
    else:
        null_index.append(df.columns[i+4])
df = df.drop(columns=null_index)
col_str = list(df.columns[0:4]) + Country_Code
df.columns = col_str
print(Country_Code)
#print(df.to_markdown())
```

```python
df
```

```python
Subregion = list(df[' Subregion'].unique())
print(Subregion)
Subregion.remove(Subregion[0])
Subregion
```

```python
Year_str=[ str(i+2018) for i in range(7)]
for j in range(len(Subregion)):
    index = list(df[(df[' Subregion'] == Subregion[j])].index)
    #print(index%7)
    for i in range(len(index)-7):
        col_name = Subregion[j]+ df.iloc[i+index[0]+7][' Country Code']+Year_str[i%7]
        df[col_name] = 0
        df.loc[index[0]+i%7, col_name ] = 1
        df.loc[i+index[0]+7, col_name ] = 1
```

```python
big_region = ['Developing Asia','Developing Asia excluding the PRC']
boo = list(df[' Country Code'].isnull())

for i in range(len(df.index)-14):
    if boo[i+14]:
        continue;
    col_name = big_region[0]+ df.iloc[i+14][' Country Code']+Year_str[i%7]
    #print(a)
    df[col_name] = 0
    df.loc[i%7, col_name ] = 1
    df.loc[i+14, col_name ] =1
col_name = big_region[0]
df[col_name] = 0
for i in range(7):
    df.loc[i%7, col_name ] = 1

for i in range(len(df.index)-14):
    if boo[i+14] or df.loc[i+14, ' Country Code']=='CHN':
        continue;
    col_name = big_region[1]+ df.iloc[i+14][' Country Code']+Year_str[i%7]
    #print(a)
    df[col_name] = 0
    df.loc[7+i%7, col_name ] = 1
    df.loc[i+14, col_name ] =1
col_name = big_region[1]
df[col_name] = 0
for i in range(7):
    df.loc[7+i%7, col_name ] = 1
df
```

```python
df = df.drop(columns=[' Subregion', ' Country Code'])
df
```

```python
print(df.columns.isnull().any().sum())

test = df[(df['Year'] > 4.5) & (df['Year'] < 6.5)]
df = df.drop(test.index)
inf = df['Inflation']
df = df.drop(columns='Inflation')
df
```

```python
test_inf = test['Inflation']
test = test.drop(columns='Inflation')
test
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
model = RandomForestRegressor(n_estimators=100, random_state=0)
# Bundle preprocessing and modeling code in a pipeline
my_pipeline = Pipeline(steps=[('model', model)])

# Preprocessing of training data, fit model
my_pipeline.fit(df, inf)

# Preprocessing of validation data, get predictions
preds = my_pipeline.predict(test)

# Evaluate the model
score = mean_absolute_error(test_inf, preds)
print('MAE:', score)
```
